```python
#part2
#spacing
from matplotlib import dates
```

```python
ax=df['Close'].plot(xlim=['2017-01-01','2017-03-01'],ylim=[50,60],figsize=(12,5))
ax.set(xlabel='')

ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))#spacing
#formatting
ax.xaxis.set_major_formatter(dates.DateFormatter('%a-%B-%d'))

ax.xaxis.set_minor_locator(dates.MonthLocator())#spacing
ax.xaxis.set_minor_formatter(dates.DateFormatter('\n\n\n\n%b'))

ax.yaxis.grid(True)
ax.xaxis.grid(True)
```
